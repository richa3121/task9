```python
import pandas as pd
import matplotlib.pyplot as plt
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Fast-load the real dataset via a raw URL to bypass your upload issues
url = "https://raw.githubusercontent.com/nsethi31/Kaggle-Data-Credit-Card-Fraud-Detection/master/creditcard.csv"
df = pd.read_csv(url)

# 1. Quick check of fraud vs non-fraud [cite: 1]
print(df['Class'].value_counts())

# 2. Prep data [cite: 2]
X = df.drop(['Time', 'Class'], axis=1)
y = df['Class']

# 3. Stratified split [cite: 3]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 4. Train Random Forest (n_estimators=100) [cite: 5]
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

# 5. Evaluate [cite: 6]
print(classification_report(y_test, rf.predict(X_test)))

# 6. Feature Importance Plot [cite: 7]

plt.figure(figsize=(10,6))
pd.Series(rf.feature_importances_, index=X.columns).nlargest(10).plot(kind='barh', color='purple')
plt.title('Real Fraud Indicators')
plt.savefig('feature_importance.png')
plt.show()

# 7. Save Model
joblib.dump(rf, 'credit_card_fraud_model.pkl')
```
